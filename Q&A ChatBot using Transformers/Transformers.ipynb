```python
from sentence_transformers import SentenceTransformer, util
```

```python
import torch
```

```python
documents = ['External Affairs Minister S Jaishankar said that while a lot of countries are nervous after Donald Trump won the US presidential elections, India is not one of them.',

'Addressing the Aditya Birla 25th Silver Jubilee Scholarship Program in Mumbai, he said that Prime Minister Narendra Modi had built rapport with multiple US presidents.',

'When he first visited Washington D.C., Barack Obama was the President, then it was Donald Trump, and then it was Joe Biden... There is something natural in terms of how he (PM Modi) forges those relationships... I know many countries are nervous about the US today, but let’s be honest. We are not one of them, he said.'
]
```

```python
documents
```

```python
len(documents)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2') #('paraphrase-MiniLM-L6-v2')
```

```python
documents_embedding = model.encode(documents)
```

```python
documents_embedding.shape
```

```python
new_text = 'who is financial minister'
```

```python
new_embedding = model.encode(new_text)
```

```python
new_embedding.shape
```

```python
cos_score = util.pytorch_cos_sim(new_embedding, documents_embedding)
cos_score
```

```python
top_result = torch.topk(cos_score , k=1)
top_result
```

```python
idx = top_result.indices[0]
idx
```

```python
documents[idx]
```

```python
def sentiment_search_engine(new_text, model):
  new_embedding = model.encode(new_text)
  cos_score = util.pytorch_cos_sim(new_embedding, documents_embedding)
  top_result = torch.topk(cos_score,  k=1)
  idx = top_result.indices[0]
  print(documents[idx])
```

```python
new_sentence1 = "Tell me about India"
new_sentence2 = "Tell me about Aditya bila"
new_sentence3 = "Tell me about modi"



sentiment_search_engine(new_sentence1, model)
```

```python

sentiment_search_engine(new_sentence2, model)
```

```python
sentiment_search_engine(new_sentence3, model)
```
